```python
# Генератор нужен тогда, когда цикл падает под серьезной нагрузкой или просто работает медленно. 
def countdown():
    i=5
    while i > 0:
        yield i
        i -= 1

for i in countdown():
    print(i)
```

```python
# Обычно это происходит от большого объема данных
def ostatokFunc (dataframe):
    i=0
    while i < len(dataframe):
        ostatok = dataframe["sepal_length"][i]%2 #Очень сложная операция, требующая много ресурсов (но тут получаем остаток от деления на 2 в столбце lan по строке с индексом == i)
        yield ostatok #возвращает значение и ставит цикл на паузу
        i += 1
        
# чтоб обработать всь датасет генератором, нужно генератор вызвать в цикле
i = 0
for each in ostatokFunc(dt): # для каждого элемента внутри того, что вернет генератор ostatokFunc
    dt["moreColumn"][i] = each #заменяем стандартное значение на то, что вернул генератор.
    i+=1
    
```
